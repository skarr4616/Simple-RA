```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def plot(df, fanout):
    
    df_numpy = df.to_numpy()

    b_tree = df_numpy[:, 0]
    heap = df_numpy[:, 1]
    
    key1, count1 = np.unique(b_tree, return_counts=True)
    key2, count2 = np.unique(heap, return_counts=True)

    key1 = [str(i) for i in key1]
    key2 = [str(i) for i in key2]

    plt.bar(key1, count1, label='B+ Tree + Heap')
    plt.bar(key2, count2, label='Heap')
    plt.legend()
    plt.xlabel('Number of Block Accesses')
    plt.ylabel('Frequency')
    plt.title('Distribution of Block Accesses for FANOUT = ' + str(fanout))
    plt.savefig('fanout_' + str(fanout) + '.png')
```

```python
path = './fanout_3.csv'
df = pd.read_csv(path)
plot(df, 3)
```

```python
path = './fanout_5.csv'
df = pd.read_csv(path)
plot(df, 5)
```
